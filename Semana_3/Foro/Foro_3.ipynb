# CONTADOR INTERACTIVO

A continuación, presento el proyecto del contador interactivo. Este proyecto consiste en una página web que muestra un número que empieza en 0 y tiene tres botones: uno para incrementar el número, otro para disminuirlo (sin que baje de 0), y otro para reiniciar el contador a su valor original. Para el diseño utilicé HTML para la estructura, CSS para los estilos y JavaScript para darle funcionalidad a los botones, permitiendo que el número cambie en tiempo real cuando se presionan.

# JavaScript

```js
const counter  = document.getElementById("counter")
const increase  = document.getElementById("increase")
const decrease  = document.getElementById("decrease")
const reset  = document.getElementById("reset")

let number = 0;

increase.addEventListener("click", () => {
    number++;
    counter.innerHTML = number;
});

decrease.addEventListener("click", () => {
    if(number==0){}
    else{
        number--;
        counter.innerHTML = number;
    }
});

reset.addEventListener("click", () => {
    number = 0;
    counter.innerHTML = number;
});

```
